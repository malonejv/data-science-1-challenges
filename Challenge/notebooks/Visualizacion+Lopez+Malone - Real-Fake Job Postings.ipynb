```python
df = df.drop('Unnamed: 0', axis='columns')
df = df.dropna(subset=['artists'])
df.info()
```

```python
df.head()
```

```python
plt.style.use('fast')

plt.figure(figsize=(8, 6))
counts = df['fraudulent'].value_counts()
bars = plt.bar(counts.index, counts.values, color=['blue', 'orange'])

# Configurar etiquetas y título
plt.xlabel('Fraudulent')
plt.ylabel('Count')
plt.title('Counts of Fraudulent vs Non-Fraudulent')
plt.xticks(counts.index, ['Non-Fraudulent', 'Fraudulent'])
plt.legend(bars, ['Non-Fraudulent', 'Fraudulent'])

plt.show()
```

```python
fraudulent_data = df[df['fraudulent'] == 1]
non_fraudulent_data = df[df['fraudulent'] == 0]

# Crear gráficos de barras separados
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
non_fraudulent_data['employment_type'].value_counts().plot(kind='bar', color='orange')
plt.title('Pub. legítimas por tipo de empleo')
plt.xlabel('Tipo de empleo')
plt.ylabel('Cantidad')

plt.subplot(1, 2, 2)
fraudulent_data['employment_type'].value_counts().plot(kind='bar')
plt.title('Pub. fraudulentas por tipo de empleo')
plt.xlabel('Tipo de empleo')
plt.ylabel('Cantidad')

# Ajustar el espacio entre los gráficos
plt.tight_layout()
```
